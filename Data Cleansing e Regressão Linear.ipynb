```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('usedf.csv')
```

```python
df.head()
```

```python
cidades=['Rio de Janeiro','São Paulo','Porto Alegre','Curitiba','Belo Horizonte','Florianópolis']  

df.describe()
```

```python
df['cidades']=df['nome_ponto']
for i in range(len(cidades)):
    for item in df['cidades']:
        if cidades[i] in item:
            df['cidades']=df['cidades'].replace(item,cidades[i])
```

```python
df['cidades'].value_counts()
```

```python
types1=["subway_station","train_station","bus_station","restaurant","bakery","supermarket","liquor_store","drugstore","shopping_mall","gym","tourist_attraction","museum","night_club"]
```

```python
df['types1']=df['types']
for i in range(len(types1)):
    for item in df['types1']:
        if types1[i] in item:
            df['types1']=df['types1'].replace(item,types1[i])
```

```python
df.head()
```

```python
classificação2 = df.groupby(['types1'])[['rating','user_ratings_total']].mean().sort_values(by=['rating'], ascending=False)
```
